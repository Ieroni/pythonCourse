2.1. Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv("kc_house_data.csv")
df.head(2)
```

```python
import matplotlib.pyplot as plt
data = df['waterfront'].value_counts()
plt.pie(data.values, autopct='%1.2f%%', labels=data.index)
plt.title("График")
plt.legend("Наличие вида на набережную")
```

2.2. Изучите распределение этажей домов

```python
plt.figure(figsize=(12,5))
sns.histplot(df['floors'], bins=15)
plt.title("распределение этажей")
plt.xlabel("floors")
plt.ylabel("houses number")
```

2.3. Изучите распределение состояния домов

```python
plt.figure(figsize=(12,5))
sns.histplot(df['grade'], bins=20)
plt.title("распределение состояни")
plt.xlabel("grade")
plt.ylabel("houses number")
```
